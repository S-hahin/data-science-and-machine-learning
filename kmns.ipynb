```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

dict={'BMI':[33.6,26.6,23.4,43.1,35.3,35.9,36.7,25.7,23.3,31],
     'Age':[50,30,40,67,23,67,45,46,29,56],
     'Sugar':[1,0,0,0,1,1,1,0,0,1]}
df=pd.DataFrame(dict)
print("details are \n",df)
x=df[['BMI','Age']]
#y=df.Sugar

km=KMeans(n_clusters=4)
km.fit(x)
pred=km.predict(x)
print(pred)

#plotting

plt.scatter(x.loc[pred==0],x.loc[pred==0],c='blue',label='cluster 1')
plt.scatter(x.loc[pred==1],x.loc[pred==1],c='red',label='cluster 2')
plt.scatter(x.loc[pred==2],x.loc[pred==2],c='green',label='cluster 3')
plt.scatter(x.loc[pred==3],x.loc[pred==3],c='yellow',label='cluster 4')
plt.legend()
plt.show()
```
